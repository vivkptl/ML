```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
user = pd.read_csv(r"user.csv")
movies = pd.read_csv(r"movies.csv")
```

```python
df = user.merge(movies)
```

```python
df.drop("timestamp", axis=1, inplace=True)
```

```python
df.head()
```

```python
df.groupby("title")["rating"].mean().sort_values(ascending=False)
```

```python
dummy = (df["title"].value_counts()>100).reset_index()
```

```python
dummy
```

```python
df = df[df["title"].isin(dummy[dummy["count"]==True]["title"])]
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df["title"].value_counts()
```

```python
df.groupby("title")["rating"].mean().sort_values(ascending=False)
```

```python
df
```

```python
user_matrix = pd.pivot_table(df, columns="user_id", index="title", values="rating", fill_value=0)
```

```python
user_corr = user_matrix.corr()
```

```python
user_corr[0].sort_values(ascending=False).head(6).tail()
```

```python
movie_matrix = pd.pivot_table(df, index="user_id", columns="title", values="rating", fill_value=0)
```

```python
movie_matrix
```

```python
movies_corr = movie_matrix.corr()
```
